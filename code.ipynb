# Dimensional Reduction using Single Value Decomposition

```python
# Reading HD Data 2 Dataset
data1 = pd.read_csv(r'C:\Users\sujal\Desktop\MA515group-16\HD Data 2.csv')
```

```python
# Printing First 5 rows of the data set
data1.head()
```

```python
# Dropped column 1 (Representing serial number)
data1 = data1.drop(columns=['Unnamed: 0'])
data1.head()
```

```python
#Size of the Data Set
data1.shape
```

```python
X = data1

#Reducing the Data Size to 8 Columns 

svd = TruncatedSVD(n_components = 8)

#The new Dataset X_svd contains the new 8 columns 
X_svd = svd.fit_transform(X)

# Naming the columns and printing the final Dimensionally reduced SVD 
columns_svd = [f'SVD column {i+1}' for i in range (8)]
X_svd_df = pd.DataFrame(data = X_svd, columns = columns_svd)

X_svd_df
```

```python
#Reducing the Data Size to 10 Columns 

svd = TruncatedSVD(n_components = 10)

#The new Dataset X_svd contains the new 10 columns 
X2_svd = svd.fit_transform(X)

# Naming the columns and printing the final Dimensionally reduced SVD 
columns_svd = [f'SVD column {i+1}' for i in range (10)]
X2_svd_df = pd.DataFrame(data = X2_svd, columns = columns_svd)

X2_svd_df
```

```python
#Reducing the Data Size to 9 Columns 
svd = TruncatedSVD(n_components = 9)

#The new Dataset X_svd contains the new 9 columns 
X3_svd = svd.fit_transform(X)

# Naming the columns and printing the final Dimensionally reduced SVD 
columns_svd = [f'SVD column {i+1}' for i in range (9)]
data3 = pd.DataFrame(data = X3_svd, columns = columns_svd)

data3
```

```python
import statsmodels.api as sm

# Function to create Q-Q plots for all columns
def qq_plots(data3):
    num_cols = data3.shape[1]
    num_rows = (num_cols + 2) // 3 
    fig, axes = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, num_rows * 3))

    for i, column in enumerate(data3.columns):
        row_idx = i // 3
        col_idx = i % 3
        ax = axes[row_idx, col_idx] if num_rows > 1 else axes[col_idx]  # Handle single-row case
        sm.qqplot(data3[column], line='45', fit=True, ax=ax)
        ax.set_title(f'Q-Q Plot - {column}')


    plt.tight_layout()
    plt.show()

# Create Q-Q plots 
qq_plots(data3)
```

```python

# Function to create pair plot with subplots in three rows

def pair_plot(data3):
    num_cols = data3.shape[1]
    num_rows = (num_cols + 2) // 3  
    fig, axes = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, num_rows * 5))

    
    for i, column in enumerate(data3.columns):
        row_idx = i // 3
        col_idx = i % 3
        ax = axes[row_idx, col_idx] if num_rows > 1 else axes[col_idx]  # Handle single-row case
        sns.histplot(data3[column], kde=True, ax=ax)
        ax.set_title(f'Histogram - {column}')

   

    plt.tight_layout()
    plt.show()

# Create pair plot
pair_plot(data3)
```

# Data Pre Processing

```python
#Separate the X and y from dataset.
X = data2.iloc[:,1:].values
y = data2.iloc[:,0].values
print('X')
print(X)
print('y')
print(y)
```

```python
# Converting Object string of ShelveLoc, Urban, US to numbers

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

labenc = LabelEncoder()
X[:,-1] = labenc.fit_transform(X[:,-1])
X[:,-2] = labenc.fit_transform(X[:,-2])
X[:,5] = labenc.fit_transform(X[:,5])

print(X)
```

```python
# Column Transformation on ShelveLoc, Urban, US transformed attributes
ct = ColumnTransformer([('encoder', OneHotEncoder(), [-1,-2,5])],remainder='passthrough')
X = np.array(ct.fit_transform(X))

print(X)
```

# Predicting Sales

```python
# Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

```

```python
# Splitting data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)

reg = LinearRegression()
```

```python
# Fitting the data using Linear Regression model
reg.fit(X_train, y_train)

# Predicting y on test dataset
y_pred = reg.predict(X_test)
```

```python
print(f'Rows in train set : { len(X_train )}\nRows in test set : {len(X_test)}\n')
```

```python
# Predicted values of sales
print(y_pred)
```

```python
# Actual sales value in test dataset
print(y_test)
```

```python
# Regression coefficient
reg.coef_
```

```python
# Intercept value
reg.intercept_
```

```python
# Regression score
reg.score(X_test, y_test)
```

# Lasso Regression

```python
# Lasso Regression with different lambdas
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split

lambdas = (0.001, 0.01, 0.1, 0.5, 1, 2, 10)
for ind, i in enumerate(lambdas):  
    reg = Lasso(alpha = i)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 4)
    reg.fit(X_train,y_train)
    y_pred = reg.predict(X_test)
    print("\nLambda value :", i)
    print("Regression Coefficients\t", reg.coef_)
    print("Regression score", reg.score(X_test, y_test))
```
